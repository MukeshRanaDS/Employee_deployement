```python
# Librearies
import pandas as pd
import numpy as np
```

```python
## Few Settings

# Ignoring warnings
import warnings 
warnings.filterwarnings('ignore')

# Displaying complete rows and columns
pd.options.display.max_columns = 100
pd.options.display.max_rows = 100
```

```python
# Data importing 
raw_df = pd.read_csv("Employee-Attrition.csv")

# Printing top 5 records
raw_df.head()
```

```python
df = raw_df.copy()
df.head()
```

```python
# Droppin unnecessary vcolumns

df.drop(['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber'], axis=1, inplace=True)
```

```python
# Creating list of Categorical Columns
cat_col = df.select_dtypes('object').columns
cat_col
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder

label_df = df.copy()
for col in cat_col:
    label_df[col] = LabelEncoder().fit_transform(label_df[col])
```

```python
# Re-Checking
label_df.head()
```

```python
top_predictors = ['Attrition','YearsAtCompany', 'StockOptionLevel', 'YearsWithCurrManager', 'Age',
       'MonthlyIncome', 'YearsInCurrentRole', 'MaritalStatus', 'JobLevel',
       'TotalWorkingYears', 'OverTime']

label_df = label_df[top_predictors]
```

```python
label_df.head()
```

```python
X = label_df.drop('Attrition', 1)
y = label_df['Attrition']
```

```python
# Checking for class distribution in target column
label_df['Attrition'].value_counts()
```

```python
# Importing library

from imblearn.combine import SMOTETomek

smk = SMOTETomek(random_state=42)
X_new, y_new = smk.fit_resample(X, y)

X_new.shape, y_new.shape
```

```python
X_new.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_new, y_new, test_size = 0.3, random_state = 0)
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = LGBMClassifier()

params = {
    'loss' : ['auto', 'binary_crossentropy'],
    'learning_rate' : [0.0001, 0.001, 0.01, 0.1, 0.5, 1],
    'max_iter' : [50, 100, 200, 300, 400, 500],
    'max_depth' : list(range(1,11)),
    'min_samples_leaf' : list(range(5,50)),
    'max_bins' : list(range(200,400))
}

random_search = RandomizedSearchCV(model, param_distributions=params, n_iter=10, 
                                   scoring='roc_auc', n_jobs=-1, cv=5, verbose=3)
random_search.fit(X_new, y_new)
```

```python
model = random_search.best_estimator_
model.fit(X_train, y_train)
```

```python
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

# Deployement

```python
input_data = (0,    0,    0,   37, 2090,    0,    2,    1,    7,    1)
input_data = np.asarray(input_data).reshape(1,-1)
prediction = model.predict(input_data)

if prediction[0]==0:
    print('Not Attrited')
elif prediction[0]==1:
    print('Attrited')
```
